```python
def process_descriptions_table_data(projects, all_time_entries, start_date, end_date):
    # Convierte la lista de entradas de tiempo y proyectos en DataFrames
    slot_entries_df = pd.DataFrame(all_time_entries)
    projects_df = pd.DataFrame(projects)

    # Resto de tu código preexistente para preprocesar los DataFrames...
    
    # Define el intervalo de fechas y conviértelo a UTC
    tz = pytz.timezone('UTC')
    start_date = pd.to_datetime(start_date).tz_localize(None).tz_localize(tz)
    end_date = pd.to_datetime(end_date).tz_localize(None).tz_localize(tz)
    
    # Filtra los datos por el intervalo temporal deseado
    mask = (slot_entries_df['slot_entrie_start'] >= start_date) & (slot_entries_df['slot_entrie_stop'] <= end_date)
    filtered_slot_entries_df = slot_entries_df.loc[mask]
    
    # Merge con projects_df
    merged_df = filtered_slot_entries_df.merge(projects_df, left_on='slot_entrie_pid', right_on='project_id', how='left')
    print(merged_df)
    # Rellena los valores NaN en la columna 'project_name' con 'Other'
    merged_df['project_name'] = merged_df['project_name'].fillna('Other projects')

    # Calcula el tiempo total para el intervalo de tiempo seleccionado
    total_time = merged_df['total_hours'].sum()

    # Crea un DataFrame con las columnas 'project_name', 'slot_entrie_description' y 'total_hours'
    summary_df = merged_df[['project_name', 'slot_entrie_description', 'total_hours']]

    # Agrupa por 'project_name' y 'slot_entrie_description' y suma 'total_hours'
    summary_df = summary_df.groupby(['project_name', 'slot_entrie_description']).sum().reset_index()

    # Calcula el porcentaje de tiempo dedicado a cada tarea
    summary_df['percentage'] = (summary_df['total_hours'] / total_time) * 100

    # Ordena el DataFrame por 'percentage' en orden descendente
    summary_df = summary_df.sort_values(by='percentage', ascending=False)

    return summary_df  # Devuelve el nuevo DataFrame
```
